```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv

html = urlopen("https://projects.fivethirtyeight.com/soccer-predictions/liga-mx/")
soup = BeautifulSoup(html, "html.parser")
```

```python
locales = soup.findAll("tr", {"class":"match-top"})
#print(locales)
for local in locales:
    print(local.get_text())
```

```python
for child in soup.find("tr", {"class":"match-top"}).children:
    print(child)
```

```python
date = soup.find("div", {"class":"match-container"}).table.tbody.tr.td.div.get_text()
print(date)

local = soup.find("div", {"class":"match-container"}).table.tbody.tr.td.next_sibling.div.next_sibling.div.get_text()
print(local)

winlocalprob = soup.find("div", {"class":"match-container"}).table.tbody.tr.td.next_sibling.next_sibling.get_text()
print(winlocalprob)

drawprob = soup.find("div", {"class":"match-container"}).table.tbody.tr.td.next_sibling.next_sibling.next_sibling.div.get_text()
print(drawprob)

visitor = soup.find("div", {"class":"match-container"}).table.tbody.tr.next_sibling.td.div.next_sibling.span.get_text()
print(visitor)

visitwinprob = soup.find("div", {"class":"match-container"}).table.tbody.tr.next_sibling.td.next_sibling.get_text()
print(visitwinprob)
```

```python
# Create a file to write to, add headers row
f = csv.writer(open("matches.csv", 'w'))
f.writerow(['Date', 'Local', 'Visitor', 'Local Win Prob', 'Visit Win Prob', 'Draw Prob'])
```
