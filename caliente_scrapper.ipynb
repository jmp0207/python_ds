```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
#utilizamos pandas y numpy para manejar los datos de los juegos de forma más fácil
import pandas as pd
import numpy as np
import requests
import json
import re
import csv

html = urlopen("https://sports.caliente.mx/es_MX/Liga-MX")
soup = BeautifulSoup(html, "html.parser")
```

```python
#Verificando el HTML que extrae
print(soup.prettify)
```

```python
#Los partidos se encuentran en esa tabla

matches = soup.find("table")

print(matches)
```

```python
#Otra forma de encontrar la referencia de la tabla
#En general, en caliente.mx sólo hay una tabla que trae todos los partidos y la información

table_matches = soup.find("table", {"class":"coupon coupon-horizontal pager-content coupon-scoreboard video-enabled"})
print(table_matches)

#variables para el loop sobre las tr que representan los juegos
teams = []
odds = []
datos_juego_row = []
datos_matrix = []
juegos_final = []
```

```python
#Cleaning de arreglos para pruebas previas
teams.clear()
odds.clear()
datos_juego_row.clear()
datos_matrix.clear()
```

```python
#Contador para loopear las celdas de las filas con los datos
i = 0 

#Loop sobre las filas que contienen los datos de los juegos
for match in table_matches.find_all("tr"):
    #print(match)
    time = match.find("span", {"class":"time"}).get_text()
    date = match.find("span", {"class":"date"}).get_text()
    
    #Loop sobre los td porque tienen la misma clase seln
    for team in match.findAll("td", {"class":"seln "}):
        teams.append(team.find("span", {"class":"seln-name"}).get_text())
        odds.append(team.find("span", {"class":"price us"}).get_text())        
    
    draw_odd = match.find("td", {"class":"seln seln_sort-D"}).find("span", {"class":"price us"}).get_text()    
    
    print(teams)
    print(odds)
    
    datos_juego_row.append(date)
    datos_juego_row.append(time)
    datos_juego_row.append(teams[i])
    datos_juego_row.append(teams[i+1])
    datos_juego_row.append(odds[i])
    datos_juego_row.append(odds[i+1])
    datos_juego_row.append(draw_odd)
    
    datos_matrix.append(datos_juego_row)
    
    i += 1
    
    

data = {'api_dev_key':"probando", 
        'api_option':'paste', 
        'api_paste_code':"prueba", 
        'api_paste_format':'python',
        'data': json.dumps(datos_matrix)} 

r = requests.post(url = "http://quehayporaqui.com/apuestas/service/save_momio", data = data)



print(r)
print(data)
print(datos_juego_row)
print(datos_matrix)
```
